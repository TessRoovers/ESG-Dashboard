```python
import pandas as pd
from sklearn.feature_extraction import text
stop = text.ENGLISH_STOP_WORDS
import enchant
from collections import Counter
import warnings
warnings.filterwarnings('ignore')
d = enchant.Dict("en_US")
words = pd.read_csv('words.csv', index_col=0)
```

```python

freqdf = pd.DataFrame(columns=["Year", "Frequency", "Country", "Document"])
for row in range(words.__len__()):
    r = words.loc[row]
    w = r["Words"]
    w = w.replace("'", '').replace("[", '').replace("]", '').replace(",", '').split()
    rw = []
    for i in w:
        if d.check(i) and len(i) > 1 and i.lower() not in stop and i.isalpha():
            rw.append(i)
    counts = Counter(rw)
    freqdf = freqdf.append({"Year":r["Year"], "Frequency":counts.most_common(20), "Country":r["Country"], "Document":r["Document"]}, ignore_index=True)
    print(f"{row}/{words.__len__()}", end="\r")

```

```python
freqdf.to_csv("freqdf.csv")
```

```python
groupedwordsdf = words.groupby(["Year", "Country"]).sum()
```

```python
groupedwordsdf.reset_index(inplace=True)
groupedwordsdf
```

```python
groupedfreqdf = pd.DataFrame(columns=["Year", "Frequency", "Country", "Document"])
for row in range(groupedwordsdf.__len__()):
    r = groupedwordsdf.loc[row]
    w = r["Words"]
    w = w.replace("'", '').replace("[", '').replace("]", '').replace(",", '').split()
    rw = []
    for i in w:
        i = i.lower()
        if d.check(i) and len(i) > 1 and i not in stop and i.isalpha():
            rw.append(i)
    counts = Counter(rw)
    groupedfreqdf = groupedfreqdf.append({"Year":r["Year"], "Frequency":counts.most_common(20), "Country":r["Country"], "Document":r["Document"]}, ignore_index=True)
    print(f"{row}/{groupedwordsdf.__len__()}", end="\r")
```

```python
groupedfreqdf.drop(groupedfreqdf.index[groupedfreqdf['Country'] == "PENDING"], inplace = True)
groupedfreqdf.to_csv("groupedfreqdf.csv")
```
